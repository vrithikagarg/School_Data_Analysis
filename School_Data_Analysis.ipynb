```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Expanded_data_with_more_features (1).csv") #change path
df.head()
df.describe()
df.info()
df.isnull().sum()
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.grid(True)
plt.show()
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Students's Score")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

# Extract true scores from the DataFrame
true_math_scores = df['MathScore'].values
true_reading_scores = df['ReadingScore'].values
true_writing_scores = df['WritingScore'].values

# Calculate the mean of the scores as the predicted value
mean_predicted_math_score = np.mean(true_math_scores)
mean_predicted_reading_score = np.mean(true_reading_scores)
mean_predicted_writing_score = np.mean(true_writing_scores)

# Create arrays of mean predicted scores with the same length as true scores
predicted_math_scores = np.full_like(true_math_scores, mean_predicted_math_score)
predicted_reading_scores = np.full_like(true_reading_scores, mean_predicted_reading_score)
predicted_writing_scores = np.full_like(true_writing_scores, mean_predicted_writing_score)

# Calculate accuracy for Math, Reading, and Writing scores using mean predicted scores
accuracy_math = accuracy_score(true_math_scores, predicted_math_scores)
accuracy_reading = accuracy_score(true_reading_scores, predicted_reading_scores)
accuracy_writing = accuracy_score(true_writing_scores, predicted_writing_scores)

print("Accuracy using Mean Predicted Math Score:", accuracy_math)
print("Accuracy using Mean Predicted Reading Score:", accuracy_reading)
print("Accuracy using Mean Predicted Writing Score:", accuracy_writing)
```
